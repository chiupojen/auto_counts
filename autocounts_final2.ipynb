```python
import pandas as pd
```

```python
# 匯入要自動計算問卷滿意度結果的csv檔案
filename = input("請輸入匯入csv檔案名稱：")
filename_csv = filename + '.csv'
print(filename_csv)
file = pd.read_csv(filename_csv)
```

```python
# 列出檔案中所有欄位名稱以及其排序號碼，以便下一步選取要計算的欄位
file_col = file.columns
i = 0
for name in file_col:
    print("column "+str(i)+" = "+str(name))
    i = i+1
```

```python
# 依據五大構面各別輸入要計入的題目編號
goal_a = int(input("請輸入課程目標構面題目起始編號："))
goal_b = int(input("請輸入課程目標構面題目結束編號："))
goal_ind = list(range(goal_a, goal_b + 1))
while True:
    print("課程目標構面題目有", goal_ind)
    del_goal = input('請輸入要刪除的題目(Enter結束):')
    if del_goal == "":
        break
    del_goal = int(del_goal)
    dg = goal_ind.count(del_goal)
    if dg > 0:
        goal_ind.remove(del_goal)
    else:
        print(del_goal, "不在課程目標構面題目中！")
teacher_a = int(input("請輸入講師表現構面題目起始編號："))
teacher_b = int(input("請輸入講師表現構面題目結束編號："))
teacher_ind = list(range(teacher_a, teacher_b + 1))
while True:
    print("講師表現構面題目有", teacher_ind)
    del_t = input('請輸入要刪除的題目(Enter結束):')
    if del_t == "":
        break
    del_t = int(del_t)
    dt = teacher_ind.count(del_t)
    if dt > 0:
        teacher_ind.remove(del_t)
    else:
        print(del_t, "不在講師表現構面題目中！")
work_a = int(input("請輸入工作助益構面題目起始編號："))
work_b = int(input("請輸入工作助益構面題目結束編號："))
work_ind = list(range(work_a, work_b + 1))
while True:
    print("工作助益構面題目有", work_ind)
    del_w = input('請輸入要刪除的題目(Enter結束):')
    if del_w == "":
        break
    del_w = int(del_w)
    dw = work_ind.count(del_w)
    if dw > 0:
        work_ind.remove(del_w)
    else:
        print(del_w, "不在工作助益構面題目中！")
env_a = int(input("請輸入設備環境構面題目起始編號："))
env_b = int(input("請輸入設備環境構面題目結束編號："))
env_ind = list(range(env_a, env_b + 1))
while True:
    print("設備環境構面題目有", env_ind)
    del_e = input('請輸入要刪除的題目(Enter結束):')
    if del_e == "":
        break
    del_e = int(del_e)
    de = env_ind.count(del_e)
    if de > 0:
        env_ind.remove(del_e)
    else:
        print(del_e, "不在設備環境構面題目中！")
sche_a = int(input("請輸入時程安排構面題目起始編號："))
sche_b = int(input("請輸入時程安排構面題目結束編號："))
sche_ind = list(range(sche_a, sche_b + 1))
while True:
    print("時程安排構面題目有", sche_ind)
    del_s = input('請輸入要刪除的題目(Enter結束):')
    if del_s == "":
        break
    del_s = int(del_s)
    ds = sche_ind.count(del_s)
    if ds > 0:
        sche_ind.remove(del_s)
    else:
        print(del_s, "不在時程安排構面題目中！")
print("課程目標構面題目有", goal_ind)
print("講師表現構面題目有", teacher_ind)
print("工作助益構面題目有", work_ind)
print("設備環境構面題目有", env_ind)
print("時程安排構面題目有", sche_ind)
```

```python
# 計算各題目各題項的回答次數
col_count = ["非常同意","同意","無意見","不同意","非常不同意"]
goal = file.iloc[: ,goal_ind]
goal_pd = []
for i in goal.keys():
    goal_count = {"題項":i}
    for j in col_count:
        goal_count[j] = goal[goal[i] == j][i].count()
    goal_pd.append(goal_count)
goal_pd_t = pd.DataFrame(goal_pd)
teacher = file.iloc[: ,teacher_ind]
teacher_pd = []
for i in teacher.keys():
    teacher_count = {"題項":i}
    for j in col_count:
        teacher_count[j] = teacher[teacher[i] == j][i].count()
    teacher_pd.append(teacher_count)
teacher_pd_t = pd.DataFrame(teacher_pd)
work = file.iloc[: ,work_ind]
work_pd = []
for i in work.keys():
    work_count = {"題項":i}
    for j in col_count:
        work_count[j] = work[work[i] == j][i].count()
    work_pd.append(work_count)
work_pd_t = pd.DataFrame(work_pd)
env = file.iloc[: ,env_ind]
env_pd = []
for i in env.keys():
    env_count = {"題項":i}
    for j in col_count:
        env_count[j] = env[env[i] == j][i].count()
    env_pd.append(env_count)
env_pd_t = pd.DataFrame(env_pd)
sche = file.iloc[: ,sche_ind]
sche_pd = []
for i in sche.keys():
    sche_count = {"題項":i}
    for j in col_count:
        sche_count[j] = sche[sche[i] == j][i].count()
    sche_pd.append(sche_count)
sche_pd_t = pd.DataFrame(sche_pd)
all_pd = [goal_pd_t, teacher_pd_t, work_pd_t, env_pd_t, sche_pd_t]
all_summary_df = pd.concat(all_pd, ignore_index=True)
display(all_summary_df)
```

```python
# 計算滿意度百分比
construct = ["課程目標", "講師表現", "工作助益", "設備環境", "時程安排"]
goal_total = goal_pd_t.sum(1).sum()
teacher_total = teacher_pd_t.sum(1).sum()
work_total = work_pd_t.sum(1).sum()
env_total = env_pd_t.sum(1).sum()
sche_total = sche_pd_t.sum(1).sum()
ans_total = [goal_total, teacher_total, work_total, env_total, sche_total]
cons_summary = []
for n in range(0,5):
    total = {"構面":construct[n]}
    for m in range(0,5):
        # print(col_count[m], all_pd[n][col_count[m]].sum(), ans_total[n])
        total[col_count[m]] = all_pd[n][col_count[m]].sum() / ans_total[n]
    cons_summary.append(total)
cons_summary_df = pd.DataFrame(cons_summary)
cons_summary_df["課程滿意度"] = cons_summary_df["非常同意"] + cons_summary_df["同意"]
display(cons_summary_df)
```

```python
# 匯出為excel檔案
output = filename + ".xlsx"
writer = pd.ExcelWriter(output)
all_summary_df.to_excel(writer, "counts")
cons_summary_df.to_excel(writer, "percentage")
writer.save()
```
